```python
file_cnt = []
for i in range(len(real_path_list)):
    folder_path = real_path_list[i]
    if Path(folder_path).exists():
        file_list = os.listdir(folder_path)
        file_count = len(file_list)
        file_cnt.append(file_count)
    else:
        file_cnt.append(0)
# len(file_cnt)
```

```python
real_path_list = df['JUPYTER_PATH'].tolist()
description_ls = df['DESCRIPTION'].tolist()
file_cnt = df['FILE_CNT'].tolist()
dcm_cnt = []
for i in range(len(real_path_list)):
    folder_path = real_path_list[i]
    file_list = os.listdir(folder_path)
    cnt = 0
    for j in range(len(file_list)):
        if file_cnt[i] == '1':
            cnt = 1
        else:
            read_name = folder_path+'/'+file_list[j]
            ds = pydicom.read_file(read_name)
            if 'StudyDescription' in dir(ds):
                if description_ls[i] == ds.StudyDescription:
                    cnt += 1
    dcm_cnt.append(cnt)
```
